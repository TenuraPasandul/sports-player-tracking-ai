```python
"""
DS5216 - Player Tracking in Sports Videos
Complete Implementation for Google Colab
"""

# ============================================================================
# STEP 1: Install Required Libraries
# ============================================================================
print("Installing required libraries...")
!pip install ultralytics opencv-python-headless yt-dlp matplotlib pillow

# ============================================================================
# STEP 2: Import Libraries
# ============================================================================
import cv2
import numpy as np
from ultralytics import YOLO
import matplotlib.pyplot as plt
from PIL import Image
import os
from google.colab import files
import yt_dlp

# Create directories for organization
os.makedirs('videos', exist_ok=True)
os.makedirs('outputs', exist_ok=True)
os.makedirs('screenshots', exist_ok=True)

print("✓ Libraries imported successfully!")

# ============================================================================
# STEP 3: Download Sports Videos from YouTube
# ============================================================================
def download_youtube_video(url, output_path='videos', max_duration=10):
    """
    Download a video from YouTube with duration limit
    """
    ydl_opts = {
        'format': 'best[height<=720]',  # 720p quality
        'outtmpl': f'{output_path}/%(title)s.%(ext)s',
        'quiet': False,
    }

    try:
        with yt_dlp.YoutubeDL(ydl_opts) as ydl:
            info = ydl.extract_info(url, download=True)
            video_path = ydl.prepare_filename(info)
            print(f"✓ Downloaded: {info['title']}")
            return video_path
    except Exception as e:
        print(f"✗ Error downloading video: {e}")
        return None

# Example YouTube URLs (Replace with your own sports videos)
youtube_urls = [
    "YOUR_YOUTUBE_URL_1",  # Cricket
    "YOUR_YOUTUBE_URL_2",  # Football
    "YOUR_YOUTUBE_URL_3",  # Rugby
    # Add 5-10 URLs here
]

print("\n📥 Downloading videos from YouTube...")
print("Note: Replace 'YOUR_YOUTUBE_URL_X' with actual video URLs")

# Uncomment below to download videos
# video_paths = []
# for url in youtube_urls:
#     if url.startswith("YOUR_"):
#         print(f"⚠ Skipping placeholder URL: {url}")
#         continue
#     path = download_youtube_video(url)
#     if path:
#         video_paths.append(path)

# ============================================================================
# ALTERNATIVE: Upload Videos Manually
# ============================================================================
print("\n📤 Alternative: Upload your videos manually")
print("Run this cell to upload videos:")
# uploaded = files.upload()
# for filename in uploaded.keys():
#     with open(f'videos/{filename}', 'wb') as f:
#         f.write(uploaded[filename])

# ============================================================================
# STEP 4: Load YOLO Model for Player Detection
# ============================================================================
print("\n🤖 Loading YOLO model...")

# Load YOLOv8 model (will auto-download on first run)
model = YOLO('yolov8n.pt')  # nano version (fastest)
# model = YOLO('yolov8s.pt')  # small version (more accurate)
# model = YOLO('yolov8m.pt')  # medium version (even more accurate)

print("✓ YOLO model loaded successfully!")

# ============================================================================
# STEP 5: Player Detection Function
# ============================================================================
def detect_players_in_video(video_path, output_path=None, conf_threshold=0.5):
    """
    Detect players (persons) in video and save annotated video

    Args:
        video_path: Path to input video
        output_path: Path to save output video
        conf_threshold: Confidence threshold for detection
    """
    if not os.path.exists(video_path):
        print(f"✗ Video not found: {video_path}")
        return None

    cap = cv2.VideoCapture(video_path)

    # Get video properties
    width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
    height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
    fps = int(cap.get(cv2.CAP_PROP_FPS))
    total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))

    print(f"\n📹 Processing: {video_path}")
    print(f"   Resolution: {width}x{height}")
    print(f"   FPS: {fps}")
    print(f"   Total Frames: {total_frames}")

    # Set output path
    if output_path is None:
        video_name = os.path.basename(video_path)
        output_path = f'outputs/detected_{video_name}'

    # Video writer
    fourcc = cv2.VideoWriter_fourcc(*'mp4v')
    out = cv2.VideoWriter(output_path, fourcc, fps, (width, height))

    frame_count = 0
    detection_results = []

    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break

        # Run YOLO detection
        results = model(frame, conf=conf_threshold, classes=[0])  # class 0 = person

        # Get detections
        boxes = results[0].boxes
        num_players = len(boxes)

        # Store results
        detection_results.append({
            'frame': frame_count,
            'num_players': num_players,
            'boxes': boxes.xyxy.cpu().numpy() if len(boxes) > 0 else [],
            'confidences': boxes.conf.cpu().numpy() if len(boxes) > 0 else []
        })

        # Annotate frame
        annotated_frame = results[0].plot()

        # Add frame info
        cv2.putText(annotated_frame, f'Frame: {frame_count} | Players: {num_players}',
                    (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 0.8, (0, 255, 0), 2)

        out.write(annotated_frame)

        # Save screenshots (every 30 frames)
        if frame_count % 30 == 0:
            screenshot_path = f'screenshots/frame_{frame_count}_{os.path.basename(video_path)}.jpg'
            cv2.imwrite(screenshot_path, annotated_frame)

        frame_count += 1

        if frame_count % 30 == 0:
            print(f"   Processed {frame_count}/{total_frames} frames...")

    cap.release()
    out.release()

    print(f"✓ Detection complete! Output saved to: {output_path}")
    print(f"✓ Screenshots saved to: screenshots/")

    return detection_results, output_path

# ============================================================================
# STEP 6: Player Tracking Function
# ============================================================================
def track_players_in_video(video_path, output_path=None, conf_threshold=0.5):
    """
    Track players across frames using YOLO tracking
    """
    if not os.path.exists(video_path):
        print(f"✗ Video not found: {video_path}")
        return None

    cap = cv2.VideoCapture(video_path)

    # Get video properties
    width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
    height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
    fps = int(cap.get(cv2.CAP_PROP_FPS))
    total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))

    print(f"\n🎯 Tracking: {video_path}")
    print(f"   Resolution: {width}x{height}")

    # Set output path
    if output_path is None:
        video_name = os.path.basename(video_path)
        output_path = f'outputs/tracked_{video_name}'

    # Video writer
    fourcc = cv2.VideoWriter_fourcc(*'mp4v')
    out = cv2.VideoWriter(output_path, fourcc, fps, (width, height))

    frame_count = 0
    tracking_results = []

    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break

        # Run YOLO tracking (tracks objects across frames)
        results = model.track(frame, conf=conf_threshold, classes=[0], persist=True)

        # Get tracked objects
        boxes = results[0].boxes

        if boxes.id is not None:
            track_ids = boxes.id.cpu().numpy().astype(int)
            num_tracked = len(track_ids)
        else:
            track_ids = []
            num_tracked = 0

        # Store results
        tracking_results.append({
            'frame': frame_count,
            'num_tracked': num_tracked,
            'track_ids': track_ids
        })

        # Annotate frame
        annotated_frame = results[0].plot()

        # Add tracking info
        cv2.putText(annotated_frame, f'Frame: {frame_count} | Tracked Players: {num_tracked}',
                    (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 0.8, (0, 255, 255), 2)

        out.write(annotated_frame)

        frame_count += 1

        if frame_count % 30 == 0:
            print(f"   Tracked {frame_count}/{total_frames} frames...")

    cap.release()
    out.release()

    print(f"✓ Tracking complete! Output saved to: {output_path}")

    return tracking_results, output_path

# ============================================================================
# STEP 7: BONUS - Keypoint Detection (Pose Estimation)
# ============================================================================
print("\n🦴 Loading YOLO Pose model for keypoint detection...")
pose_model = YOLO('yolov8n-pose.pt')  # Pose estimation model
print("✓ Pose model loaded!")

def detect_keypoints_in_video(video_path, output_path=None, conf_threshold=0.5):
    """
    Detect player keypoints (body joints) in video
    """
    if not os.path.exists(video_path):
        print(f"✗ Video not found: {video_path}")
        return None

    cap = cv2.VideoCapture(video_path)

    # Get video properties
    width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
    height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
    fps = int(cap.get(cv2.CAP_PROP_FPS))
    total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))

    print(f"\n🦴 Keypoint Detection: {video_path}")

    # Set output path
    if output_path is None:
        video_name = os.path.basename(video_path)
        output_path = f'outputs/keypoints_{video_name}'

    # Video writer
    fourcc = cv2.VideoWriter_fourcc(*'mp4v')
    out = cv2.VideoWriter(output_path, fourcc, fps, (width, height))

    frame_count = 0

    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break

        # Run pose estimation
        results = pose_model(frame, conf=conf_threshold)

        # Annotate frame with keypoints
        annotated_frame = results[0].plot()

        # Add info
        cv2.putText(annotated_frame, f'Frame: {frame_count} | Keypoint Detection',
                    (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 0.8, (255, 0, 255), 2)

        out.write(annotated_frame)
        frame_count += 1

        if frame_count % 30 == 0:
            print(f"   Processed {frame_count}/{total_frames} frames...")

    cap.release()
    out.release()

    print(f"✓ Keypoint detection complete! Output: {output_path}")

    return output_path

# ============================================================================
# STEP 8: Performance Metrics Calculation
# ============================================================================
def calculate_metrics(detection_results):
    """
    Calculate performance metrics
    """
    total_detections = sum([len(r['boxes']) for r in detection_results])
    avg_confidence = np.mean([np.mean(r['confidences'])
                             for r in detection_results if len(r['confidences']) > 0])
    avg_players_per_frame = np.mean([r['num_players'] for r in detection_results])

    metrics = {
        'total_frames': len(detection_results),
        'total_detections': total_detections,
        'avg_confidence': avg_confidence,
        'avg_players_per_frame': avg_players_per_frame
    }

    return metrics

# ============================================================================
# STEP 9: Visualization Functions
# ============================================================================
def visualize_detection_stats(detection_results, video_name):
    """
    Create visualization plots for the report
    """
    frames = [r['frame'] for r in detection_results]
    num_players = [r['num_players'] for r in detection_results]

    plt.figure(figsize=(12, 6))

    # Plot 1: Number of players per frame
    plt.subplot(1, 2, 1)
    plt.plot(frames, num_players, linewidth=2)
    plt.xlabel('Frame Number')
    plt.ylabel('Number of Players Detected')
    plt.title(f'Player Detection - {video_name}')
    plt.grid(True, alpha=0.3)

    # Plot 2: Confidence distribution
    plt.subplot(1, 2, 2)
    all_confidences = []
    for r in detection_results:
        if len(r['confidences']) > 0:
            all_confidences.extend(r['confidences'])

    if all_confidences:
        plt.hist(all_confidences, bins=30, edgecolor='black')
        plt.xlabel('Confidence Score')
        plt.ylabel('Frequency')
        plt.title('Detection Confidence Distribution')
        plt.grid(True, alpha=0.3)

    plt.tight_layout()
    plt.savefig(f'outputs/stats_{video_name}.png', dpi=300, bbox_inches='tight')
    plt.show()

    print(f"✓ Statistics plot saved!")

# ============================================================================
# STEP 10: Process All Videos in the Videos Folder
# ============================================================================
def process_all_videos(video_folder='videos', conf_threshold=0.5):
    """
    Process all videos in the specified folder
    """
    # Get all video files
    video_extensions = ['.mp4', '.avi', '.mov', '.mkv', '.webm']
    video_files = []

    for file in os.listdir(video_folder):
        if any(file.lower().endswith(ext) for ext in video_extensions):
            video_files.append(os.path.join(video_folder, file))

    if not video_files:
        print("⚠ No video files found in the videos folder!")
        return

    print(f"\n📁 Found {len(video_files)} video(s) to process")
    print("="*70)

    all_results = {}

    for idx, video_path in enumerate(video_files, 1):
        video_name = os.path.basename(video_path)
        print(f"\n{'='*70}")
        print(f"Processing Video {idx}/{len(video_files)}: {video_name}")
        print(f"{'='*70}")

        try:
            # 1. Player Detection
            print("\n🔍 Step 1: Player Detection...")
            detection_results, output_video = detect_players_in_video(
                video_path, conf_threshold=conf_threshold
            )

            # 2. Player Tracking
            print("\n🎯 Step 2: Player Tracking...")
            tracking_results, tracked_video = track_players_in_video(
                video_path, conf_threshold=conf_threshold
            )

            # 3. Keypoint Detection (Bonus)
            print("\n🦴 Step 3: Keypoint Detection (Bonus)...")
            keypoint_video = detect_keypoints_in_video(
                video_path, conf_threshold=conf_threshold
            )

            # 4. Calculate metrics
            print("\n📊 Step 4: Calculating Performance Metrics...")
            metrics = calculate_metrics(detection_results)

            print("\n" + "="*50)
            print(f"📊 METRICS FOR: {video_name}")
            print("="*50)
            for key, value in metrics.items():
                if isinstance(value, float):
                    print(f"   {key}: {value:.4f}")
                else:
                    print(f"   {key}: {value}")
            print("="*50)

            # 5. Visualize results
            print("\n📈 Step 5: Generating Visualizations...")
            visualize_detection_stats(detection_results, video_name.replace('.', '_'))

            # Store results
            all_results[video_name] = {
                'detection_results': detection_results,
                'tracking_results': tracking_results,
                'metrics': metrics,
                'output_video': output_video,
                'tracked_video': tracked_video,
                'keypoint_video': keypoint_video
            }

            print(f"\n✅ Successfully processed: {video_name}")

        except Exception as e:
            print(f"\n❌ Error processing {video_name}: {str(e)}")
            continue

    return all_results

# ============================================================================
# STEP 11: Generate Summary Report
# ============================================================================
def generate_summary_report(all_results):
    """
    Generate a comprehensive summary report
    """
    print("\n" + "="*70)
    print("📋 SUMMARY REPORT - ALL VIDEOS")
    print("="*70)

    for video_name, results in all_results.items():
        metrics = results['metrics']
        print(f"\n🎬 Video: {video_name}")
        print("-" * 50)
        print(f"   Total Frames: {metrics['total_frames']}")
        print(f"   Total Detections: {metrics['total_detections']}")
        print(f"   Avg Confidence: {metrics['avg_confidence']:.4f}")
        print(f"   Avg Players/Frame: {metrics['avg_players_per_frame']:.2f}")
        print(f"   Output Files:")
        print(f"      - Detection: {results['output_video']}")
        print(f"      - Tracking: {results['tracked_video']}")
        print(f"      - Keypoints: {results['keypoint_video']}")

    print("\n" + "="*70)
    print("✅ All processing complete!")
    print(f"📁 Check the following folders for outputs:")
    print(f"   - outputs/ : Processed videos")
    print(f"   - screenshots/ : Frame captures")
    print("="*70)

# ============================================================================
# STEP 12: Main Execution
# ============================================================================
print("\n" + "="*70)
print("🚀 PLAYER TRACKING SYSTEM READY!")
print("="*70)

# List existing videos
print("\n📁 Checking existing videos...")
if os.path.exists('videos'):
    video_files = [f for f in os.listdir('videos')
                   if f.endswith(('.mp4', '.avi', '.mov', '.mkv', '.webm'))]
    if video_files:
        print(f"✓ Found {len(video_files)} video(s):")
        for i, vf in enumerate(video_files, 1):
            print(f"   {i}. {vf}")
    else:
        print("⚠ No videos found in 'videos' folder")
else:
    print("⚠ 'videos' folder not found")

# Check YOLO models
print("\n🤖 Checking YOLO models...")
model_files = [f for f in os.listdir('.') if f.endswith('.pt')]
if model_files:
    print(f"✓ Found {len(model_files)} YOLO model(s):")
    for mf in model_files:
        print(f"   - {mf}")

print("\n" + "="*70)
print("📝 QUICK START - Run this code to process all videos:")
print("="*70)
print("""
# Process all videos at once
all_results = process_all_videos(video_folder='videos', conf_threshold=0.5)

# Generate summary report
generate_summary_report(all_results)

# Download all output videos (optional)
# from google.colab import files
# for video_name, results in all_results.items():
#     files.download(results['output_video'])
#     files.download(results['tracked_video'])
#     files.download(results['keypoint_video'])
""")
# Process all videos at once
all_results = process_all_videos(video_folder='videos', conf_threshold=0.5)

# Generate summary report
generate_summary_report(all_results)
#from google.colab import files
#for video_name, results in all_results.items():
#    files.download(results['output_video'])
#    files.download(results['tracked_video'])
#    files.download(results['keypoint_video'])

print("\n" + "="*70)
print("💡 OR process a single video:")
print("="*70)
print("""
# Single video processing
video_path = 'videos/your_video.mp4'  # Change this to your video name

# Run all steps
detection_results, output_video = detect_players_in_video(video_path, conf_threshold=0.5)
tracking_results, tracked_video = track_players_in_video(video_path, conf_threshold=0.5)
keypoint_video = detect_keypoints_in_video(video_path, conf_threshold=0.5)

# Show metrics
metrics = calculate_metrics(detection_results)
print("\\n📊 Metrics:", metrics)

# Visualize
visualize_detection_stats(detection_results, 'my_video')
""")

print("\n✨ Ready to process! Run the code above to start.")

# AUTO-RUN: Process all videos automatically (uncomment to enable)
# print("\n🚀 Auto-processing all videos...")
# all_results = process_all_videos(video_folder='videos', conf_threshold=0.5)
# generate_summary_report(all_results)
```
